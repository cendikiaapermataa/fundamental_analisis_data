## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** "Kategori produk apa saja yang menyumbang total pendapatan (revenue) tertinggi dan terendah di platform e-commerce ini sepanjang tahun 2018, dan bagaimana informasi ini dapat memandu strategi alokasi anggaran promosi oleh tim marketing?"

Keterangan:
- **Specific**: Fokus analisis sangat jelas, yaitu mengidentifikasi kategori produk dengan kontribusi pendapatan paling besar dan paling kecil di platform e-commerce, serta mengaitkannya dengan strategi promosi.
- **Measurable**: Total pendapatan dapat diukur dengan angka pasti melalui penjumlahan nilai transaksi (kolom price) untuk masing-masing kategori produk.
- **Action-Oriented**: Tim marketing dapat langsung menggunakan insight ini untuk mengambil tindakan nyata, seperti mengoptimalkan anggaran promosi pada kategori terlaris untuk memaksimalkan margin, atau mengevaluasi ulang strategi untuk kategori dengan pendapatan terendah.
- **Relevant**: Mengetahui kategori mana yang menjadi sumber pendapatan utama sangat krusial dan relevan dengan tujuan bisnis e-commerce untuk menjaga profitabilitas perusahaan
- **Time-bound**: Analisis dibatasi pada rentang waktu yang spesifik, yaitu transaksi yang terjadi "sepanjang tahun 2018".


- **Pertanyaan 2:** "Bagaimana tren persentase keterlambatan pengiriman pesanan pada setiap bulannya selama tahun 2018, dan bagaimana hubungan antara keterlambatan pengiriman tersebut dengan skor ulasan (review score) dari pelanggan?

Keterangan:
- **Specific**: Fokus pada eksplorasi tren bulanan dan pengamatan korelasi antara dua variabel operasional: waktu pengiriman (terlambat vs tepat waktu) dan tingkat kepuasan pelanggan (skor ulasan).
- **Measurable**: Keterlambatan diukur secara objektif dengan membandingkan tanggal estimasi pengiriman dan tanggal aktual barang diterima oleh pelanggan. Dampaknya diukur dengan rata-rata review score (skala 1-5).
- **Action-Oriented**: Jika terbukti keterlambatan menurunkan kepuasan pelanggan secara signifikan pada bulan-bulan tertentu, tim operasional tahu bahwa mereka harus segera mengevaluasi kinerja mitra logistik untuk mencegah churn.
- **Relevant**: Pengiriman yang tepat waktu adalah salah satu Key Performance Indicator (KPI) utama yang menentukan retensi dan loyalitas pelanggan di industri e-commerce.
- **Time-bound**: Dibatasi rentang waktunya secara spesifik, yaitu pengamatan tren bulanan "selama tahun 2018".

## Data Wrangling

### Assessing Data

#### Identifying ... problem

```python
# Membuat list dataframe beserta namanya
datasets = {
    "Customers": customers_df,
    "Geolocation": geolocation_df,
    "Order Items": order_items_df,
    "Order Payments": order_payments_df,
    "Order Reviews": order_reviews_df,
    "Orders": orders_df,
    "Product Category":product_category_df,
    "Products": products_df,
    "Sellers": sellers_df
}

# Looping untuk mengecek duplikat dan missing values
for name, df in datasets.items():
    print(f"--- Info Data: {name} ---")
    print(f"Jumlah baris: {df.shape[0]}")
    print(f"Jumlah duplikat: {df.duplicated().sum()}")
    print(f"Jumlah Missing Values:\n{df.isnull().sum()}\n")

# Mengecek tipe data secara spesifik untuk tabel yang akan sering digunakan (orders dan order_reviews)
print("=== Info Tipe Data Orders ===")
orders_df.info()
print("\n=== Info Tipe Data Order Reviews ===")
order_reviews_df.info()
```

**Steps to Take:**
- Menangani Missing Value: - Pada orders_df, menghapus (drop) baris data yang nilai order_delivered_customer_date-nya kosong, karena analisis waktu pengiriman hanya valid untuk pesanan yang benar-benar sampai ke pelanggan.
    - Pada order_reviews_df, mengisi (impute) kekosongan pada review_comment_title dan review_comment_message dengan teks "No Comment" agar struktur data tetap utuh tanpa membuang nilai review score.
    - Pada products_df, mengisi product_category_name yang kosong dengan kategori "Unknown" agar data tersebut tetap bisa dihitung dalam analisis pendapatan.

- Menangani Duplicate Data: - Pada geolocation_df, menghapus seluruh baris data yang duplikat menggunakan metode drop_duplicates agar representasi titik lokasi menjadi unik dan akurat.

- Memperbaiki Tipe Data (Formatting): - Mengubah tipe data seluruh kolom yang merepresentasikan tanggal dan waktu pada orders_df (seperti order_purchase_timestamp) dan order_reviews_df (seperti review_creation_date) dari tipe object (string) menjadi tipe datetime agar dapat dikalkulasi.

**Insight:** (Opsional)
- Adanya missing value pada kolom tanggal pengiriman di dataset pesanan (orders) sebenarnya bukanlah sebuah "kecacatan" atau error sistem, melainkan hal yang sangat logis dalam basis data operasional. Kekosongan ini merepresentasikan status pesanan yang sedang dalam perjalanan (processing/shipped) atau pesanan yang dibatalkan (canceled).

- Tingginya persentase missing value pada kolom komentar teks di dataset ulasan (order_reviews) menunjukkan karakteristik umum konsumen e-commerce, di mana mayoritas pembeli lebih suka memberikan penilaian instan berupa rating bintang (angka) daripada harus meluangkan waktu untuk mengetik ulasan secara tertulis.

- Ratusan ribu data duplikat pada dataset geolocation mengindikasikan bahwa sistem mencatat banyak pelanggan atau penjual yang tinggal di kodepos atau apartemen/gedung yang sama, sehingga menghasilkan titik koordinat lintang dan bujur yang identik.

### Cleaning Data

#### Fixing ... problem

```python
# 1. Membersihkan tabel orders_df:
# a. Menghapus data pesanan yang tidak/belum sampai ke pelanggan
orders_df.dropna(subset=['order_delivered_customer_date'], inplace=True)
# b. Mengubah tipe data teks menjadi datetime
datetime_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]
for column in datetime_columns:
    orders_df[column] = pd.to_datetime(orders_df[column])

# 2. Membersihkan tabel order_reviews_df:
# Mengubah tipe data waktu dan mengisi komentar kosong
order_reviews_df['review_creation_date'] = pd.to_datetime(order_reviews_df['review_creation_date'])
order_reviews_df['review_answer_timestamp'] = pd.to_datetime(order_reviews_df['review_answer_timestamp'])
order_reviews_df.fillna(value={"review_comment_message": "No Comment", "review_comment_title": "No Comment"}, inplace=True)

# 3. Membersihkan tabel geolocation_df: Menghapus duplikat
geolocation_df.drop_duplicates(inplace=True)

# 4. Membersihkan tabel products_df: Mengisi kategori yang kosong dengan 'Unknown'
products_df.fillna({'product_category_name': 'Unknown'}, inplace=True)

# Memeriksa kembali hasil pembersihan
print("--- Hasil Setelah Cleaning ---")
print(f"Sisa duplikat di Geolocation: {geolocation_df.duplicated().sum()}")
print(f"Sisa missing value di Kategori Produk: {products_df['product_category_name'].isnull().sum()}")
print(f"Sisa missing value di Waktu Tiba Pesanan: {orders_df['order_delivered_customer_date'].isnull().sum()}")
print(f"Sisa missing value di Komentar Ulasan: {order_reviews_df['review_comment_message'].isnull().sum()}")
print("\nTipe data waktu di orders_df saat ini:")
print(orders_df[datetime_columns].dtypes)
```

**Insight:** (Opsional)
- Penghapusan baris kosong pada order_delivered_customer_date membuat dataset kini difokuskan secara eksklusif pada pesanan yang benar-benar sukses dan sampai ke tangan pelanggan (bebas dari data pesanan yang batal/masih diproses). Di sisi lain, pengisian nilai (imputasi) dengan "No Comment" pada ulasan dan "Unknown" pada kategori produk berhasil menyelamatkan puluhan ribu baris data agar tetap bisa dianalisis review score dan harga transaksinya tanpa harus membuang informasi yang berharga.

- Penghapusan ratusan ribu data duplikat pada tabel geolocation membuat dataset menjadi jauh lebih ringan dan merepresentasikan titik lokasi (kodepos) pelanggan/penjual secara unik. Hal ini akan mencegah double-counting atau bias jika dilakukan pemetaan geospasial.

- Seluruh kolom yang memuat informasi tanggal dan waktu (baik di tabel pesanan maupun ulasan) kini telah terstandarisasi menjadi tipe data datetime64[ns]. Perubahan ini merupakan fondasi paling krusial agar kita bisa melakukan operasi matematis, seperti memfilter transaksi khusus di tahun 2018 dan menghitung durasi keterlambatan pengiriman pada tahap eksplorasi selanjutnya.

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# 1. Menggabungkan order_items dengan orders
orders_items_merged = pd.merge(
    left=order_items_df,
    right=orders_df,
    how="inner",
    on="order_id"
)

# 2. Menggabungkan hasil di atas dengan products
orders_products_merged = pd.merge(
    left=orders_items_merged,
    right=products_df,
    how="inner",
    on="product_id"
)

# 3. Menggabungkan dengan terjemahan kategori
all_df = pd.merge(
    left=orders_products_merged,
    right=product_category_df,
    how="left",
    on="product_category_name"
)

# 4. Menggabungkan dengan order_reviews (untuk analisis kepuasan pelanggan)
all_df = pd.merge(
    left=all_df,
    right=order_reviews_df,
    how="left",
    on="order_id"
)

# Jika ada kategori bahasa Inggris yang kosong, kita isi dengan nama aslinya
all_df['product_category_name_english'] = all_df['product_category_name_english'].fillna(all_df['product_category_name'])

# Memfilter khusus pesanan sukses (delivered) di tahun 2018 sesuai Pertanyaan Bisnis
all_df_2018 = all_df[(all_df['order_purchase_timestamp'].dt.year == 2018) &
                     (all_df['order_status'] == 'delivered')].copy()


# A. PERTANYAAN 1: Kategori Produk dengan Revenue Tertinggi & Terendah (2018)
category_revenue = all_df_2018.groupby('product_category_name_english').price.sum().sort_values(ascending=False)

print("--- Top 5 Kategori Produk dengan Revenue Tertinggi (2018) ---")
print(category_revenue.head(5))

print("\n--- Bottom 5 Kategori Produk dengan Revenue Terendah (2018) ---")
print(category_revenue.tail(5))


# B. PERTANYAAN 2: Tren Keterlambatan dan Dampak pada Review Score (2018)
# Buat kolom penanda terlambat (True jika aktual > estimasi)
all_df_2018['is_late'] = all_df_2018['order_delivered_customer_date'] > all_df_2018['order_estimated_delivery_date']
all_df_2018['month'] = all_df_2018['order_purchase_timestamp'].dt.strftime('%Y-%m')

# Tren persentase keterlambatan tiap bulan
delay_trend = (all_df_2018.groupby('month')['is_late'].mean() * 100).round(2)
print("\n--- Tren Persentase Keterlambatan Bulanan Tahun 2018 (%) ---")
print(delay_trend)

# Dampak keterlambatan terhadap skor ulasan (Review Score)
# (Drop duplikat order_id agar 1 pesanan tidak dihitung berulang rating-nya)
review_impact = all_df_2018.drop_duplicates(subset=['order_id']).groupby('is_late')['review_score'].mean().round(2)
review_impact.index = ['Tepat Waktu / Lebih Cepat', 'Terlambat']
print("\n--- Rata-rata Review Score berdasarkan Status Pengiriman ---")
print(review_impact)
```

**Insight:** (Opsional)
- Performa Pendapatan Kategori Produk (Q1): Data menunjukkan adanya ketimpangan penguasaan pasar yang sangat masif. Kategori health_beauty mendominasi mutlak sebagai tulang punggung perusahaan di tahun 2018 dengan meraup total pendapatan tertinggi mencapai 757.081, disusul ketat oleh watches_gifts. Di sisi lain, kategori niche seperti cds_dvds_musicals memiliki performa penjualan yang sangat memprihatinkan (hanya menyumbang pendapatan sebesar 65.00 sepanjang tahun). Hal ini mengindikasikan bahwa audiens di platform ini sangat terfokus pada gaya hidup dan perawatan diri, sedangkan produk media fisik/hiburan klasik hampir tidak memiliki peminat.

- Fluktuasi Keterlambatan Logistik (Q2): Tren bulanan mengungkap adanya indikasi "krisis logistik" pada kuartal pertama 2018. Tingkat pesanan yang terlambat melonjak drastis secara beruntun dari 6.53% di bulan Januari hingga mencapai titik puncaknya pada Maret 2018 (20.45% pesanan terlambat). Meskipun demikian, terdapat indikasi perbaikan operasional yang sangat baik setelahnya, di mana angka keterlambatan berhasil ditekan secara signifikan hingga menyentuh titik terendahnya pada Juni 2018 (hanya 1.34%).

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# 1. Mengubah Series menjadi DataFrame agar mudah divisualisasikan dengan Seaborn
category_revenue_df = category_revenue.reset_index()
category_revenue_df.columns = ['product_category_name_english', 'price']

# 2. Menyiapkan data untuk visualisasi (Top 5 dan Bottom 5)
top_5_revenue = category_revenue_df.head(5)
# Karena sekarang sudah jadi DataFrame, parameter 'by' sudah bisa digunakan
bottom_5_revenue = category_revenue_df.tail(5).sort_values(by='price', ascending=True)

# 3. Membuat figure matplotlib
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

# Warna kustom agar estetis
colors_top = ["#2E86AB"] + ["#D3D3D3"] * 4 # Warna biru untuk peringkat 1, abu-abu sisanya
colors_bottom = ["#D90429"] + ["#D3D3D3"] * 4 # Warna merah untuk peringkat terbawah

# Plot Top 5 (Kiri)
sns.barplot(x="price", y="product_category_name_english", data=top_5_revenue, palette=colors_top, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel("Total Pendapatan (BRL)", fontsize=12)
ax[0].set_title("5 Kategori Produk Pendapatan Tertinggi (2018)", loc="center", fontsize=15, pad=15)
ax[0].tick_params(axis ='y', labelsize=12)

# Plot Bottom 5 (Kanan)
sns.barplot(x="price", y="product_category_name_english", data=bottom_5_revenue, palette=colors_bottom, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel("Total Pendapatan (BRL)", fontsize=12)
ax[1].invert_xaxis() # Membalik sumbu x agar grafik terlihat simetris berhadap-hadapan
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("5 Kategori Produk Pendapatan Terendah (2018)", loc="center", fontsize=15, pad=15)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Perbandingan Performa Kategori Produk Berdasarkan Pendapatan", fontsize=20, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
# Membuat figure dengan 2 subplot (Atas dan Bawah)
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(14, 12))

# --- PLOT 1: Tren Persentase Keterlambatan (Garis) ---
# Ubah index bulan menjadi kolom agar bisa di-plot
delay_trend_df = delay_trend.reset_index()

sns.lineplot(
    x="month",
    y="is_late",
    data=delay_trend_df,
    marker="o",
    linewidth=3,
    markersize=8,
    color="#F4A261", # Warna jingga estetis
    ax=ax[0]
)
ax[0].set_title("Tren Persentase Keterlambatan Pengiriman Bulanan (Tahun 2018)", loc="center", fontsize=16, pad=55)
ax[0].set_xlabel("Bulan", fontsize=12)
ax[0].set_ylabel("Persentase Terlambat (%)", fontsize=12)
ax[0].tick_params(axis='x', labelsize=11)
ax[0].tick_params(axis='y', labelsize=11)

# Menambahkan anotasi angka pada titik tertinggi (Maret 2018)
max_delay_val = delay_trend_df['is_late'].max()
max_delay_idx = delay_trend_df[delay_trend_df['is_late'] == max_delay_val].index[0]
ax[0].annotate(f'Puncak Keterlambatan\n({max_delay_val}%)',
             xy=(max_delay_idx, max_delay_val),
             xytext=(max_delay_idx, max_delay_val + 2),
             ha='center',
             arrowprops=dict(facecolor='red', shrink=0.05))


# --- PLOT 2: Dampak Keterlambatan terhadap Review Score (Batang) ---
# Ubah index status pengiriman menjadi kolom
review_impact_df = review_impact.reset_index()
review_impact_df.columns = ['Status Pengiriman', 'Rata-rata Review Score']

colors_review = ["#2A9D8F", "#E76F51"] # Hijau untuk tepat waktu, Merah untuk telat

# Menggunakan parameter hue dan legend=False agar terhindar dari warning
sns.barplot(
    x="Status Pengiriman",
    y="Rata-rata Review Score",
    hue="Status Pengiriman",
    data=review_impact_df,
    palette=colors_review,
    legend=False,
    ax=ax[1]
)
ax[1].set_title("Dampak Keterlambatan terhadap Kepuasan Pelanggan (Review Score)", loc="center", fontsize=16, pad=15)
ax[1].set_xlabel(None)
ax[1].set_ylabel("Rata-rata Skor (1-5)", fontsize=12)
ax[1].set_ylim(0, 5) # Maksimal skor 5
ax[1].tick_params(axis='x', labelsize=12)
ax[1].tick_params(axis='y', labelsize=11)

# Menambahkan angka tepat di atas batang agar mudah dibaca
for p in ax[1].patches:
    ax[1].annotate(format(p.get_height(), '.2f'),
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'center',
                   xytext = (0, 9),
                   textcoords = 'offset points',
                   fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

Secara visual, grafik ini menjadi bukti tak terbantahkan bahwa keterlambatan pengiriman adalah "pembunuh" kepuasan pelanggan (customer satisfaction).

## Analisis Lanjutan (Opsional)

```python
# 1. Menggabungkan all_df dengan customers_df untuk mendapatkan 'customer_unique_id'
rfm_df = pd.merge(
    left=all_df,
    right=customers_df[['customer_id', 'customer_unique_id']],
    how="inner",
    on="customer_id"
)

# 2. Menentukan tanggal referensi (Anggap "hari ini" adalah tanggal transaksi terakhir di dataset + 1 hari)
recent_date = rfm_df['order_purchase_timestamp'].max() + dt.timedelta(days=1)

# 3. Menghitung Recency, Frequency, dan Monetary untuk setiap pelanggan
rfm_summary = rfm_df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (recent_date - x.max()).days, # Recency: Selisih hari sejak transaksi terakhir
    'order_id': 'nunique',                                              # Frequency: Jumlah transaksi unik
    'price': 'sum'                                                      # Monetary: Total uang yang dihabiskan
}).reset_index()

# 4. Mengganti nama kolom agar lebih intuitif
rfm_summary.columns = ['customer_unique_id', 'Recency', 'Frequency', 'Monetary']

# Tampilkan sampel pelanggan dengan nilai Monetary tertinggi
print("--- Top 5 Pelanggan berdasarkan Total Belanja (Monetary) ---")
print(rfm_summary.sort_values(by='Monetary', ascending=False).head())
print("\n" + "="*60 + "\n")

# VISUALISASI DISTRIBUSI RFM
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(22, 6))

# Menggunakan palet warna earth tone yang konsisten
colors_rfm = ["#2E86AB", "#2A9D8F", "#E76F51"]

# Plot Recency
sns.histplot(rfm_summary['Recency'], bins=50, color=colors_rfm[0], ax=ax[0], kde=True)
ax[0].set_title('Distribusi Recency\n(Hari Sejak Transaksi Terakhir)', fontsize=14, pad=10)
ax[0].set_xlabel('Recency (Hari)')
ax[0].set_ylabel('Jumlah Pelanggan')

# Plot Frequency
# (Mayoritas pelanggan e-commerce hanya beli 1x, kita filter <= 5 agar grafik mudah dibaca)
sns.countplot(x='Frequency', data=rfm_summary[rfm_summary['Frequency'] <= 5], color=colors_rfm[1], ax=ax[1])
ax[1].set_title('Distribusi Frequency\n(Jumlah Transaksi per Pelanggan)', fontsize=14, pad=10)
ax[1].set_xlabel('Frequency (Jumlah Transaksi)')
ax[1].set_ylabel('Jumlah Pelanggan')

# Plot Monetary
# (Kita batasi total belanja <= 2000 agar grafik tidak terdistorsi oleh outlier/sultan)
sns.histplot(rfm_summary[rfm_summary['Monetary'] <= 2000]['Monetary'], bins=50, color=colors_rfm[2], ax=ax[2], kde=True)
ax[2].set_title('Distribusi Monetary\n(Total Belanja <= 2000 BRL)', fontsize=14, pad=10)
ax[2].set_xlabel('Monetary (Total BRL)')
ax[2].set_ylabel('Jumlah Pelanggan')

plt.suptitle("Analisis RFM: Pemahaman Perilaku Retensi Pelanggan")
plt.tight_layout()
plt.show()
```

**Insight:** (Opsional)

Melalui metode RFM (Recency, Frequency, Monetary), kita dapat mengekstrak pola perilaku dan loyalitas pengguna secara lebih mendalam:

- Recency (Rentang Waktu Belanja): Mayoritas pelanggan memiliki rentang waktu transaksi terakhir (recency) yang tersebar cukup merata, namun terdapat puncak (peak) yang signifikan pada kurun waktu 100 hingga 250 hari terakhir. Ini menunjukkan bahwa platform cukup berhasil mempertahankan arus transaksi yang stabil dalam beberapa bulan terakhir.

- Frequency (Tingkat Pembelian Berulang): Sifat transaksi di platform ini sangat didominasi oleh "pembeli sekali jalan" (one-time buyers). Grafik menunjukkan bahwa hampir seluruh pelanggan (lebih dari 90.000 pengguna) hanya pernah melakukan 1 kali transaksi, dan sangat jarang yang kembali untuk melakukan pembelian kedua atau ketiga.

- Monetary (Daya Beli Pelanggan): Distribusi daya beli pelanggan (uang yang dihabiskan) sangat condong ke kanan (right-skewed), di mana mayoritas pelanggan berfokus pada transaksi berskala kecil hingga menengah (menghabiskan kurang dari 250 BRL).

- Kesimpulan Analisis Lanjutan: Platform e-commerce ini memiliki kinerja yang sangat luar biasa dalam melakukan akuisisi (menarik pengguna baru untuk bertransaksi), namun memiliki kendala besar dalam retensi dan membangun loyalitas pelanggan (Customer Lifetime Value yang sangat rendah). Sebagai rekomendasi tambahan, perusahaan sangat perlu membangun program loyalitas (seperti poin reward, diskon khusus member, atau voucher gratis ongkir untuk pembelian kedua) guna mengubah one-time buyer menjadi pelanggan tetap.
